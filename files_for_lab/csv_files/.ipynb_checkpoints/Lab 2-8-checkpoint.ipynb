# ROUND 2

```python
import pandas as pd
import numpy as np
data_mca = pd.read_csv('marketing_customer_analysis.csv')

# Show the dataframe shape
data_mca.shape
```

```python
# Change every header into lower case
lowercase = []
for column in data_mca.columns:
    lowercase.append(column.lower())
lowercase
data_mca.columns = lowercase

data_mca.columns

# Put a space in employmentstatus

data_mca = data_mca.rename(columns={'employmentstatus':'employment status'})

data_mca.columns

# Underscore every space 

data_mca.columns = data_mca.columns.str.replace(" ", "_")
data_mca

# Drop unnamed:_0

data_mca = data_mca.drop(['unnamed:_0'], axis =1)
```

```python
# Numerical columns

data_mca.select_dtypes(np.number) # will show us how many columns are numerical 

# Categorical columns

data_mca.select_dtypes(object) # will show us how many columns are categorical
```

```python
# Check for duplicates & drop them

sum(data_mca.duplicated())

data_mca = data_mca.drop_duplicates()

# Check for null value

data_mca.isna().sum()

# Let´s check the percentage of the null on the total to decide wheter to delate or not

round(data_mca.isna().sum()/len(data_mca),3)*100

# Let´s begin with vehicle type since more than 50% of values are null

data_mca['vehicle_type'].value_counts(dropna=False) # We can see that there are only two values 
data_mca['vehicle_type']=data_mca['vehicle_type'].fillna('M') # We are going to assume that the cars that are not automatic are manual

# For the rest of columns with nan values, since they represent a small % of total, we are going to drop them

data_mca=data_mca.dropna()

# Let´s check if we have dealt with nan

data_mca.isna().sum()
```

```python
data_mca['effective_to_date']= pd.to_datetime(data_mca['effective_to_date'])
data_mca.dtypes
data_mca['month']=data_mca['effective_to_date'].dt.month
```

```python
# Show a plot of the total number of responses

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

sns.countplot(x='response', data=data_mca)
plt.ylabel('Number of responses')
plt.show()

# sns.displot(data_mca['response'], bins=40)
```

```python
# Show a plot of the response rate by the sales channel

# Let´s begin by showing a plot with the count of responses by channel

sns.countplot(x='sales_channel', hue='response', data=data_mca)
plt.ylabel('Responses by channel')
plt.show()


# Now, let´s create the rate

data_mca['response_rate'] = np.where(data_mca.response=='Yes',1,0) # This creates a new column where the yes values in response take value of 1 and the no 0
data_mca['response_rate']
data_mca.response_rate.value_counts() # There are a total of 1308 yes and 7826 no

# Use a groupby function to create the rate 

data_mca.groupby('sales_channel')['response_rate'].mean().plot.bar() # En este caso ponemos la media pq es la suma de los 1 'yes' por cada canal/de la longitud. Esta es el rate 
plt.show()
#sns.barplot(x='sales_channel', y='response_rate', data=data_mca) # Otra forma de mostrar el gráfico
```

```python
# Show a plot of the response rate by the total claim amount

# We can proceed in the same way than in the previous

# data_mca.groupby('total_claim_amount')['response_rate'].mean().plot.bar() #Debido a que la mayoría de los datos son únicos mostrar todos es inviavle. Tenemos que agruparlos

print('Maximum value', data_mca['total_claim_amount'].max())
print('Minimum value', data_mca['total_claim_amount'].min())
print('Mean value', data_mca['total_claim_amount'].mean())

# Now we are going to create a function to bin into categories

def total_claim_amount_bins(x):
    if x<=450:
        return 1
    elif x<=900:
        return 2
    elif x<= 1350:
        return 3
    elif x<= 1800:
        return 4
    else:
        return 5
    
data_mca['total_claim_amount_bins']=list(map(total_claim_amount_bins, data_mca['total_claim_amount'])) #Map ejecuta la función para cada valor de data_mca['total_claim_amount']
data_mca # En la columna total_claim_amount_bins podemos ver a que intervalo pertenece cada uno 

# Now that we have all values binned, we can create the plot

#data_mca.groupby('total_claim_amount_bins')['response_rate'].mean().plot.bar() #One representation
#plot.show()
sns.barplot(x= 'total_claim_amount_bins', y='response_rate', data=data_mca) #another representation
```

```python
# Show a plot of the response rate by income

# We can proceed in the same way than in the previous

# data_mca.groupby('income')['response_rate'].mean().plot.bar() #Debido a que la mayoría de los datos son únicos mostrar todos es inviavle. Tenemos que agruparlos

print('Maximum value', data_mca['income'].max())
print('Minimum value', data_mca['income'].min())
print('Mean value', data_mca['income'].mean())

# Now we are going to create a function to bin into categories

def income_bins(x):
    if x<=33333:
        return 1
    elif x<=66666:
        return 2
    elif x<=99999:
        return 3
    else:
        return 4
    
data_mca['income_bins']=list(map(income_bins, data_mca['income'])) #Map ejecuta la función para cada valor de data_mca['total_claim_amount']
data_mca # En la columna total_claim_amount_bins podemos ver a que intervalo pertenece cada uno 

# Now that we have all values binned, we can create the plot

#data_mca.groupby('total_claim_amount_bins')['response_rate'].mean().plot.bar() #One representation
#plot.show()
sns.barplot(x= 'income_bins', y='response_rate', data=data_mca) #another represe
```

```python
#X-y split

# total_claim_amount is the target variable

y = numerical['total_claim_amount']
X = numerical.drop(['total_claim_amount'], axis=1) # Include all the data without total_claim_amount
```

```python
#Normalize. In this case we´ll normalize wirt MinMaxScaler

# y = numerical['total_claim_amount'] #Defined previously
# X = numerical.drop(['total_claim_amount'], axis=1) #Defined previously

from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X)
x_normalized = transformer.transform(X)
                                
```

# ROUND 6

```python
# Onehot encoder categorical

from sklearn.preprocessing import OneHotEncoder
categoricals

encoder = OneHotEncoder(drop='first').fit(categoricals)

columns_encoder = encoder.get_feature_names_out(input_features=categoricals.columns)

categoricals_encoded = pd.DataFrame(encoder.transform(categoricals).toarray(), columns=columns_encoder)

categoricals_encoded.head() # The customer names are not well extracted
```

```python
# concat data frames 

X_total= pd.concat([X, categoricals_encoded], axis=1)

```

```python
# Train Test Split # When I try to test X_total I gen an error. The model will be built using x_normalized pending to check

from sklearn.model_selection import train_test_split as test

X_train, X_test, y_train, y_test=test(x_normalized, y, test_size=0.2)
```

```python
# Apply Linear Regression

from sklearn.linear_model import LinearRegression as LR

LR = LR()
LR.fit(X_train, y_train)
predictions = LR.predict(X_train)
print(predictions)
```

```python
# Model Validation

from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

mae = mean_absolute_error(y_test, predictions)
print(mae)



```
