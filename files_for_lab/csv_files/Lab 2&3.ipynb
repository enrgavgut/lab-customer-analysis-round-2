```python
# Change every header into lower case
lowercase = []
for column in data_mca.columns:
    lowercase.append(column.lower())
lowercase
data_mca.columns = lowercase

data_mca.columns

# Put a space in employmentstatus

data_mca = data_mca.rename(columns={'employmentstatus':'employment status'})

data_mca.columns

# Underscore every space 

data_mca.columns = data_mca.columns.str.replace(" ", "_")
data_mca

# Drop unnamed:_0

data_mca = data_mca.drop(['unnamed:_0'], axis =1)
```

```python
# Numerical columns

data.select_dtypes(np.number) # will show us how many columns are numerical 

# Categorical columns

data.select_dtypes(object) # will show us how many columns are categorical
```

```python
# Check for duplicates & drop them

#sum(data_mca.duplicated())

data_mca = data_mca.drop_duplicates()

# Check for null value

#data_mca.isna().sum()

# Let´s check the percentage of the null on the total to decide wheter to delate or not

#round(data_mca.isna().sum()/len(data_mca),3)*100

# strategy: drop rows that have null values and are object columns "response" and "state"
data_mca = data_mca[data_mca['state'].isna()==False]
data_mca = data_mca[data_mca['response'].isna()==False]

# strategy: include mean value in rows which are numerical columns "months_since_last_claim" and "number_of_open_complaints"

mean_months_since_last_claim = data_mca['months_since_last_claim'].mean()
mean_months_since_last_claim
data_mca['months_since_last_claim'] = data_mca['months_since_last_claim'].fillna(mean_months_since_last_claim)


mean_number_of_open_complaints = data_mca['number_of_open_complaints'].mean()
mean_number_of_open_complaints
data_mca['number_of_open_complaints'] = data_mca['number_of_open_complaints'].fillna(mean_number_of_open_complaints)

# strategy: for vehicle_class, size and type we are going to fill the null with the most frequent value (mode) 

data_mca['vehicle_class'].unique() # This shows all the values of vehicle_size in an array
data_mca['vehicle_class'].value_counts(dropna=False) # Count of the values including the NaN. The most commom is Four-Door Car
data_mca['vehicle_class'] = data_mca['vehicle_class'].fillna('Four-Door Car')

data_mca['vehicle_size'].unique() # This shows all the values of vehicle_size in an array
data_mca['vehicle_size'].value_counts(dropna=False) # Count of the values including the NaN. The most commom is Medsize
data_mca['vehicle_size'] = data_mca['vehicle_size'].fillna('Medsize')

data_mca['vehicle_type'].unique() # This shows all the values of vehicle_type in an array
data_mca['vehicle_type'].value_counts(dropna=False) # Count of the values including the NaN. 
data_mca['vehicle_type'] = data_mca['vehicle_type'].fillna('A')

# Let´s check if we have dealt with all NaN

data_mca.isna().sum()
```
